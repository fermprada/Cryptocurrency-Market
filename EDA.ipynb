CONCLUSIÓN

En los diagramas de caja podemos observar valores atipicos para cada moneda, los puntos observados fuera del rango de la caja representan anomalías lo cuál es normal entre criptomonedas ya que cualquier factor externo puede hacer que afecte el precio, su capitalización y sobre todo su volumen de mercado, por lo cual podemos decir que siempre van a ser un factor cambiante sin poder predecirlo con exactitud.

En los histogramas podemos ver que si bien la mayoría tiene un modelo de "cola derecha", ya que “se afina" a medida que nos movemos a lo largo del eje x, lo que indica que la frecuencia está disminuyendo, hay otros que por contrario la frecuencia aumenta y de pronto se disminuye de vuelta por lo que retomamos que no se sigue un modelo fijo sino que demuestra que va a ser cambiante dependiendo de como ya se mencionó, factores externos.

Por último, en los gráficos de líneas podemos observar las tendencias en el precio de cada moneda desde que fueron lanzadas hasta la actualidad, y vemos como en el pasar de los años algunas han tenido tanto picos bastante elevados como unas bajas bastante notables, lo que nos indica que de un momento a otro dependiendo de factores como por ejemplo, la oferta y demanda esto puede variar altamente y es un factor a considerar al momento de realizar inversiones con alguna de estas monedas.
